# Import Libraries & Data

```python
import numpy as np
import pandas as pd
import math
from scipy.stats import rankdata
from sklearn.utils import shuffle
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
from statistics import mean

def regression_results(y_true, y_pred):
    # Regression metrics
    explained_variance=metrics.explained_variance_score(y_true, y_pred)
    mean_absolute_error=metrics.mean_absolute_error(y_true, y_pred) 
    mse=metrics.mean_squared_error(y_true, y_pred) 
    #mean_squared_log_error=metrics.mean_squared_log_error(y_true, y_pred)
    median_absolute_error=metrics.median_absolute_error(y_true, y_pred)
    r2=metrics.r2_score(y_true, y_pred)
    print('explained_variance: ', round(explained_variance,4))    
    #print('mean_squared_log_error: ', round(mean_squared_log_error,4))
    print('r2: ', round(r2,4))
    print('MAE: ', round(mean_absolute_error,4))
    print('MSE: ', round(mse,4))
    print('RMSE: ', round(np.sqrt(mse),4))        

datasales = pd.read_csv('data/BrandTotalSales.csv')
dataunits = pd.read_csv('data/BrandTotalUnits.csv')
branddetails = pd.read_csv('data/BrandDetails.csv')
```

# Data Cleaning (Object -> datetime64 & Float64)

```python
# Datasales
#First convert our months to datetime
datasales['Months'] = pd.to_datetime(datasales['Months'])
#Total units is too large currently to convert to a float
datasales['Total Sales ($)'] = datasales['Total Sales ($)'].str[:10]
datasales['Total Sales ($)'] = datasales['Total Sales ($)'].str.replace(',', '')
datasales['Total Sales ($)'] = pd.to_numeric(datasales['Total Sales ($)'])

# Dataunit
#First convert our months to datetime
dataunits['Months'] = pd.to_datetime(dataunits['Months'])
#Total units is too large currently to convert to a float
dataunits['Total Units'] = dataunits['Total Units'].str[:10]
dataunits['Total Units'] = dataunits['Total Units'].str.replace(',', '').astype(float)
dataunits['Total Units'] = pd.to_numeric(dataunits['Total Units'])
```

# Calculating Overall Market Cap

```python
market_cap = pd.DataFrame(columns = ['Months','Market'])
month_i = dataunits['Months'].unique()
month_i.sort()
for month in month_i:
    total = datasales.loc[datasales['Months']==month,'Total Sales ($)'].sum()
    temp = {'Months': month,'Market':total}
    market_cap = market_cap.append(temp,ignore_index = True)
market_cap.loc[:,'Previous Month Market'] = market_cap.loc[:,'Market'].shift(1)
market_cap = market_cap.drop(columns='Market')
market_cap = market_cap.fillna(0)
market_cap.head()
```

# Feature Engineering for Each Brands

### 1. Merge data.  2. Time Series Feature  3. Brand Features

```python
#Iteration
final_data = pd.DataFrame()
market_share_avg_set = []
brands = dataunits["Brands"].unique()
#Time Series: Previous Month unit, rolling average unit, etc.
i=1
for brand in brands:
    if (i%100 == 0):
        print('{}th Brand Processing'.format(i))
    units = dataunits[dataunits.Brands == brand]

    units = units.assign(Previous_month_unit=units.loc[:,'Total Units'].shift(1))
    units = units.assign(Previous_Month2=units.loc[:,'Total Units'].shift(2))
    units = units.assign(Previous_Month3=units.loc[:,'Total Units'].shift(3))
    units = units.fillna(0)
    units = units.assign(Rolling_avg_unit=(units.Previous_month_unit+units.Previous_Month2+units.Previous_Month3)/3)
    units = units.drop(columns=['Previous_Month2','Previous_Month3'])

    # Month, Year, Open Days
    units = units.assign(Month = (units.Months).dt.month)
    units = units.assign(Year = (units.Months).dt.year)
    units = units.assign(Holiday_season = np.where((units.Month>=10),1,0))
    units = units.assign(Open_days = (units.Months-min(units.Months)).astype('timedelta64[D]'))

    #Merge Sales Data
    sales = datasales[datasales.Brand==brand]
    units = units.merge(sales,left_on = 'Months',right_on='Months')
    units = units.drop(columns='Brand')
    units = units.assign(Previous_month_sale=units.loc[:,'Total Sales ($)'].shift(1))
    units = units.assign(Previous_Month2=units.loc[:,'Total Sales ($)'].shift(2))
    units = units.assign(Previous_Month3=units.loc[:,'Total Sales ($)'].shift(3))
    units = units.fillna(0)
    units = units.assign(Rolling_avg_sale=(units.Previous_month_sale+units.Previous_Month2+units.Previous_Month3)/3)
    units = units.drop(columns=['Previous_Month2','Previous_Month3'])
    #ARP
    units = units.assign(previous_ARP=(units.loc[:,'Total Sales ($)']/units.loc[:,'Total Units']).shift(1))
    #Merge Market_Cap
    units = units.merge(market_cap,left_on = 'Months',right_on='Months')
    units = units.assign(previous_market_share = (units.loc[:,'Previous_month_sale']/units.loc[:,'Previous Month Market']))
    units = units.fillna(0)
    market_share_avg = units.previous_market_share.mean()
    
    #Add market_share_avg to dictionary
    market_share_avg_set.append(market_share_avg)
    
    
    #Brand Features
    details = branddetails[branddetails.Brand == brand]  
    flower = 0
    roll = 0
    candy = 0
    vape = 0
    if 'Flower' in details['Category L2'].values:
        flower = 1
    if 'Pre-Rolled' in details['Category L2'].values:
        roll = 1
    if 'Candy' in details['Category L3'].values:
        candy = 1      
    if 'Vape' in details['Category L3'].values:
        vape = 1       
    units['Flower'] = flower
    units['Pre-Rolled'] = roll
    units['Candy'] = candy
    units['Vape'] = vape
    units['ProdCount'] = details['Product Description'].count()
    units['Strain_Variety'] = len(details['Strain'].unique())
    units = units.fillna(0)
    
    # Add Feature Engineered Data for each brand to final data
    final_data = final_data.append(units)
    i=i+1
print('Feature Engineering Completed')
```

### Additional Data Feature Engineering:  Engineering the Categorical Variable: Brands

#### Instead of Hot Encoding Brands (this will create 1640 columns), rank the brands by its average market share. The brand with the lowest market share will have rank 1 and the brand with the highest market share will have rank 1640

```python
market_share_avg_set = [0 if math.isnan(x) else x for x in market_share_avg_set]
rank_list = rankdata(market_share_avg_set,method = 'min')
j = 0
for brand in brands:
    if(j%200==0):
        print('{}th Brand Processing'.format(j))
    final_data.loc[final_data['Brands']==brand,'Rank'] = rank_list[j]
    j=j+1
print('Ranking Completed')
```

# Linear Regression (R-squared & P-values)

###  Splitting Training & Test Data
#### 80% of Brands in Training Data, 20% of Brands in Test Data

```python
def train_test_split_by_brands(xi,yi,brands_list,train_i,test_i):
    
    x_train = xi.loc[xi['Brands'].isin(brands_list[train_i])]
    x_test = xi.loc[xi['Brands'].isin(brands_list[test_i])]
    y_train_pd = yi.loc[yi['Brands'].isin(brands_list[train_i])]
    y_test_pd = yi.loc[yi['Brands'].isin(brands_list[test_i])]

    x_train = x_train.drop(columns='Brands')
    x_test = x_test.drop(columns='Brands')
    return x_train, x_test, y_train_pd['Total Sales ($)'].tolist(),y_test_pd['Total Sales ($)'].tolist()
```

```python
x = final_data.drop(columns=['Months','Total Sales ($)','Total Units'])
y_sales = final_data[['Brands','Total Sales ($)']]
brands_final = final_data['Brands'].unique()
brands_shuffle = shuffle(brands_final)

test_index = range(0,int(len(brands_shuffle)*0.2))
train_index = range(int(len(brands_shuffle)*0.2)+1,len(brands_shuffle))

x_train, x_test, y_train, y_test = train_test_split_by_brands(x,y_sales,brands_shuffle,train_index,test_index)
```

```python
print('x_train size: ', x_train.shape)
print('y_train size: ', len(y_train))
print('x_test size: ', x_test.shape)
print('y_test size: ',len(y_test))
```

### Pipeline - Standardize & Fill nan with 0s

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

# Numerical - age, trestbp, cholserum, thalach, oldpeak
# Categorical - cp, restecg, slope, ca, thal
# Binary - sex, fbs, exang

numerical_features = [x for x in  x_train.columns]

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="constant",fill_value=0)),
    ('std_scaler',StandardScaler())
])

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, numerical_features)
])
x_train_piped = full_pipeline.fit_transform(x_train)
x_test_piped = full_pipeline.transform(x_test)
```

### Model Fitting (OLS)

```python
import statsmodels.api as sm
x_train_piped_sm =sm.add_constant(x_train_piped)
ols = sm.OLS(y_train,x_train_piped_sm)
result = ols.fit()
print(result.summary())
```

###  Model performance with test data

```python
x_test_piped_sm =sm.add_constant(x_test_piped)
ypred = result.predict(x_test_piped_sm)
```

```python
regression_results(y_test,ypred)
```

# Cross Validation 
### (Function defined to split data so that test and train data have no common brands)

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits = 10, shuffle = False)
```

```python
def cross_val(method,x,y,cv,brands,pipeline,trees=100,depth=10,feature=8):
    i=0
    r2_results = []
    for train_index,test_index in cv.split(brands):
        x_train, x_test, y_train, y_test = train_test_split_by_brands(x,y,brands,train_index,test_index)
        x_train_piped = pipeline.fit_transform(x_train)
        x_test_piped = pipeline.transform(x_test)
        if(method == 'OLS'):
            x_train_piped_sm = sm.add_constant(x_train_piped)
            model = sm.OLS(y_train,x_train_piped_sm)
            result = model.fit()
            x_test_piped_sm = sm.add_constant(x_test_piped)
            ypred = result.predict(x_test_piped_sm)
            r2=metrics.r2_score(y_test, ypred)
            r2_results.append(r2)
        elif(method == 'RandomForest'):
            model = RandomForestRegressor(n_estimators=trees, max_depth=depth, max_features=feature)
            result = model.fit(x_train_piped,y_train)
            ypred=result.predict(x_test_piped)
            r2 = metrics.r2_score(y_test,ypred)
            r2_results.append(r2)
         
    return r2_results
```

```python
r2_OLS = cross_val('OLS',x,y_sales,kf,brands_shuffle,full_pipeline)
r2_RF = cross_val('RandomForest',x,y_sales,kf,brands_shuffle,full_pipeline)
```

```python
print('Linear Regression Cross Validation Mean R-squared: ',mean(r2_OLS))
print('Random Forest Cross Validation Mean R-squared: ',mean(r2_RF))
```

# Grid Search - Find Optimal Parameters

```python
def cross_val_grid_search(x,y,brands,pipeline,trees_array,depth_array,features_array):
    total_split = len(trees_array)*len(depth_array)*len(features_array)
    kf = KFold(n_splits = total_split, shuffle = False)
    max_r2 = 0
    max_parameter = [0,0,0]
    i=0
    j=0
    k=0
    for train_index,test_index in kf.split(brands):
        x_train, x_test, y_train, y_test = train_test_split_by_brands(x,y,brands,train_index,test_index)
        x_train_piped = pipeline.fit_transform(x_train)
        x_test_piped = pipeline.transform(x_test)
        regr = RandomForestRegressor(n_estimators = trees_array[i], max_depth = depth_array[j], max_features = features_array[k])
        regr.fit(x_train_piped,y_train)
        y_pred = regr.predict(x_test_piped)
        r2 = metrics.r2_score(y_test,y_pred)
        if r2>max_r2:
            max_r2 = r2
            max_parameter = [trees_array[i],depth_array[j],features_array[k]]
        
        k=k+1
        if (k == len(features_array)):
            k=0
            print('[Trees:{} Depth:{} Features:{}] parameter processed'.format(trees_array[i],depth_array[j],'*'))
            j=j+1
            if(j==len(depth_array)):
                j=0
                i=i+1
        
    return max_r2,max_parameter
```

```python
trees = [100,200,400,600,800]
depths = [5,10,25,50,100]
features = [1,3,5,8,10]

max_rsqured,max_param = cross_val_grid_search(x,y_sales,brands_shuffle,full_pipeline,trees,depths,features)
```

```python
print('Max R2: ',max_rsqured)
print('Parameter [n_estimator, max_depths, max_features]: ', max_param)
```
